```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
%pip install SQLAlchemy
%pip install mysqlclient
from sqlalchemy import create_engine
```

```python
# Define the URL of the website to be scraped
url1 = 'https://www.britannica.com/topic/European-Union'
url2 = 'https://en.wikipedia.org/wiki/List_of_sovereign_states_in_Europe_by_GDP_(nominal)'

extended_request1 = requests.get(url1)
extended_request2 = requests.get(url2)
```

```python
soup1 = BeautifulSoup(extended_request1.text, 'html.parser')
soup2 = BeautifulSoup(extended_request2.text, 'html.parser')    
```

```python
print(soup1.prettify())
```

```python
print(soup2.prettify())
```

```python
# Scraped the country names and GDP values for 2018 and 2023
countries = soup1.findAll('a', class_='md-crosslink', attrs={'data-show-preview': 'true'})[2:29] 

country_names = soup2.find('table', class_='wikitable').findAll('a', attrs={'title': True})[0:49]
gdp_values_2023 = soup2.find('table', class_='wikitable').findAll('td')[3:837:17]
gdp_values_2018 = soup2.find('table', class_='wikitable').findAll('td')[8:842:17]
```

```python
# Create a dictionary to store the scraped data
eu_countries = {
    'eu_country' : []
}
# Loop through the scraped data and store in the dictionary
for country in countries:
    eu_country = country.text
    eu_countries['eu_country'].append(eu_country)
    print('EU Country: ',eu_country)
    print('-'*75)
    
```

```python
# Create a dictionary to store the scraped data
gdp_data_2023 = {
    'country_name' : [],
    'year': [],
    'gdp' : []
}
# Loop through the scraped data and store in the dictionary
for country_name, gdp_value in zip(country_names, gdp_values_2023):
    country = country_name.text
    gdp = gdp_value.text

    gdp_data_2023['country_name'].append(country)
    print('Country Name:', country)

    gdp_data_2023['year'].append(2023)
    print('Year:', 2023)

    gdp_data_2023['gdp'].append(gdp)
    print('GDP:', gdp)
    print('-' * 75)

```

```python
# Create a dictionary to store the scraped data
gdp_data_2018 = {
    'country_name' : [],
    'year': [],
    'gdp' : []
}
# Loop through the scraped data and store in the dictionary
for country_name, gdp_value in zip(country_names, gdp_values_2018):
    country = country_name.text
    gdp = gdp_value.text

    gdp_data_2018['country_name'].append(country)
    print('Country Name:', country)

    gdp_data_2018['year'].append(2018)
    print('Year:', 2018)

    gdp_data_2018['gdp'].append(gdp)
    print('GDP:', gdp)
    print('-' * 75)
```
