## What is the average Logistics Performance Index score for each EU country for the years 2018 and 2023, and how do they rank against each other?

```python
query1 = f"""
SELECT 
    c.Country, AVG(l.LPI_Score) AS Average_LPI_Score, DENSE_RANK() OVER (ORDER BY AVG(l.LPI_Score) DESC) AS 'Rank'
FROM 
    eu_countries c 
JOIN 
    lpi_data l ON c.CountryID = l.CountryID
GROUP BY 
    c.Country
ORDER BY 
    Average_LPI_Score DESC
"""
df1 = pd.read_sql(query1, engine)
```

Insight: Germany has the highest LPI score among all the countries in the EU.

Recommendation: Establish Germany as the logistics center for product distribution in the EU.

Prediction: By strategically positioning logistics center in a country with a high LPI score, overall costs related to poor shipment handing should decrease. This could lead to improve margins and customer satisfaction.

## How does the average GDP of EU countries correlate with their average LPI score for the years 2018 and 2023? 

```python
query2 = f"""
WITH AVG_LPI AS (
SELECT 
    CountryID, AVG(LPI_Score) AS Average_LPI_Score, AVG(Infrastructure_Score) AS Average_Infrastructure_Score
FROM 
    lpi_data
GROUP BY 
    CountryID
),
AVG_GDP AS (
SELECT 
    CountryID, AVG(GDP_Billions_of_USD) AS Average_GDP
FROM 
    gdp_data
GROUP BY 
    CountryID
)
SELECT  
    c.CountryID, c.Country, l.Average_LPI_Score, DENSE_RANK() OVER (ORDER BY l.Average_LPI_Score DESC) AS 'LPI_Rank', g.Average_GDP, DENSE_RANK() OVER (ORDER BY g.Average_GDP DESC) AS 'GDP_Rank'
FROM 
    eu_countries c
JOIN 
    AVG_LPI l ON c.CountryID = l.CountryID
JOIN 
    AVG_GDP g ON l.CountryID = g.CountryID
ORDER BY 
    GDP_Rank
"""
df2 = pd.read_sql(query2, engine)
```

Insight: Germany stays on top in terms of GDP and LPI.  

Recommendation: Capitalize on Germany's economy and logistics for business growth and expansion. 

Prediction: Increase presence in the EU.
