```python
# Import necessary libraries
import pandas as pd 
import requests
import os
%pip install openai
from openai import OpenAI
import json
%pip install SQLAlchemy
%pip install mysqlclient
from sqlalchemy import create_engine
```

```python
# Set the API key from secrets
openai_api_key = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(api_key=openai_api_key)
```

```python
# Read the CSV data
data = pd.read_csv('eu_countries.csv')

# List to store responses
responses1 = []

# Iterate over the country values
for country in data['eu_country']:
    response = client.completions.create(
        model="gpt-3.5-turbo-instruct",
        prompt=f"What is the capital city of {country}?",
        max_tokens=100
    )
    # Append the response text to the list
    responses1.append(response.choices[0].text)
```

```python
responses1
```

```python
# List to store responses
responses2 = []

# Iterate over the capital city values
for city in responses1:
    response = client.completions.create(
        model="gpt-3.5-turbo-instruct",
        prompt=f"What is the UN/LOCODE (United Nations Code for Trade and Transport Locations) of {city}?",
        max_tokens=100
    )
    # Append the response text to the list
    responses2.append(response.choices[0].text)
```

```python
responses2
```

```python
df = pd.DataFrame({'EU_Country': data['eu_country'], 'Capital_City': responses1, 'UN_Location_Code' : responses2}) 
```

```python
df
```

```python
# Save the data to a CSV file to then clean with Excel
df.to_csv('eu_countries_unlocode.csv', index=False)
```

```python
# Set the API key from secrets
api_key = os.getenv('FREIGHTOS_API_KEY')
# Endpoint URL for the Freight Rate Estimator API
url = 'https://api.freightos.com/api/v1/freightEstimates'

# Headers including the required API key
headers = {
    'Content-Type': 'application/json',
    'x-apikey': api_key
}

# request body for the Freight Rate Estimator API
# Only destination input is changed 1-by-1 
payload = {
  "load": [
    {
      "quantity": 1,
      "unitType": "boxes",
      "unitVolumeCBM": 0.1
    }
  ],
  "legs": [
    {
      "origin": {
        "unLocationCode": "USLAX",
        "airportCode": "LAX"
      },
      "destination": {
        "unLocationCode": "SESTO", 
        "airportCode": "STO"
      },
    }
  ]
}

response = requests.post(url, json=payload, headers=headers)
```

```python
response.status_code
```

```python
# Compile responses to one Excel file (1-by-1)
# CYNIC response unavailable
# FRPAR airportCode is CDG
# ITROM airportCode is FCO
# ROBUH airportCode is OTP
response.json()
```
